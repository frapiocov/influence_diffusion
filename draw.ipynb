```python
import snap
import math
import networkx as nx
import random
import pandas as pd
import matplotlib.pyplot as plt

net_df = pd.read_csv("data/soc-sign-bitcoinalpha.csv") # gli edge della rete

nx_graph = nx.from_pandas_edgelist(net_df, source = "source", target = "target")
pos = nx.spring_layout(nx_graph)
    
nx.draw(nx_graph, pos, with_labels=False, node_color='skyblue', edge_color='black', node_size=10)
plt.savefig("graph.png")
```
